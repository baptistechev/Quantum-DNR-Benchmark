```python
import json

benchmark_runs = []
with open('benchmark_runs.jsonl', 'r') as f:
    for line in f:
        benchmark_runs.append(json.loads(line))
```

```python
quantum_NN_2_pp_runs_15_bus = sorted(
    [run for run in benchmark_runs if run['sim_type'] == 'quantum_NN_2_pp' and run['network'] == "33_bus"],
    key=lambda x: (x['network'], x['run_index'])
)
print(f"Found {len(quantum_NN_2_pp_runs_15_bus)} runs with sim_type 'quantum_NN_2_pp' and network '15_bus'")
```

```python
best_run = min(quantum_NN_2_pp_runs_15_bus, key=lambda run: run["details"][0]["objective"])

print("run_index:", best_run["run_index"])
print("objective:", best_run["details"][0]["objective"])
print("timestamp_utc:", best_run["timestamp_utc"])
```
